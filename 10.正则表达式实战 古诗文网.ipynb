```python
import requests
import re
```

```python
# 用正则表达式来爬虫，网页中的任何都系比如<div>都是没有任何功能
# 只能被看做是字符而已
def parse_page(url):
    headers = {
        'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    text = response.text
    # re.DOTALL 可以让.匹配\n
    titles = re.findall(r'<div\sclass="cont">.*?<b>(.*?)</b>', text, re.DOTALL) #DOTALL 匹配所有的字符（包括换行）
    dynasties = re.findall(r'<p class="source">.*?<a.*?>(.*?)</a>', text, re.DOTALL)
    authors = re.findall(r'<p class="source">.*?<a.*?><a.*?>(.*?)</a>', text, re.DOTALL)
    content_tags = re.findall(r'<div class="contson" .*?">(.*?)</div>', text, re.DOTALL)
    contents = list(re.sub(r'<.*?>', "", content).strip() for content in content_tags)
    '''
    # 等同于
    contents = []
    for content in content_tags:
        x = re.sub(r'<.*?>', "", content)
        print(x.strip())
        contents.append(x.strip())
    '''
    poems = []
    for value in zip(titles, dynasties, authors, contents):
        title, dynasty, author, content = value
        poem = {
            'title': title,
            'dynasty': dynasty,
            'authro': author,
            'content': content
        }
        poems.append(poem)
    return poems
    
    
def main():
    poems = []
    for page in range(1, 101):
        url = 'https://www.gushiwen.org/default_%s.aspx' % page
        poems_in_the_page = parse_page(url)
        poems.append(poems_in_the_page)
    for poem in poems:
        print(poem)
        print('='*30)

if __name__ == '__main__':
    main()
```
